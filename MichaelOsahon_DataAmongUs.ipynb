```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('DataAmongUs.csv', newline='') as file:
    my_reader = csv.reader(file)
    headers = next(my_reader)
    my_data = []
    for row in my_reader:
        my_data.append(row)

print(headers)
print(my_data)
```

```python
age_category = {}

for datum in my_data:
    age = datum[0]
    if age in age_category:
        age_category[age] += 1
    else:
        age_category[age] = 1

print(age_category)
# Most popular age category was 22 - 25 (with 10 votes)
```

```python
data_0 = {'Age Category': ['Under 21', '22 - 25', '26 - 30'], 
          'Frequency': [1, 10, 5]}

df_0 = pd.DataFrame(data_0)

plt.pie(df_0['Frequency'], labels = df_0['Age Category'])
plt.title('Distribution of Age')
plt.show()
```

```python
music_genre = {}

for datum in my_data:
    music = datum[1]
    if music in music_genre:
        music_genre[music] += 1
    else:
        music_genre[music] = 1

print(music_genre)
# Most popular music genre was Pop with 7 votes, while Jazz & Indie both had 5 votes
```

```python
data_1 = {'Music Genre': ['Pop', 'Jazz', 'RnB', 'Afrobeats', 'Rock', 'Indie',], 
          'Frequency': [7, 5, 3, 3, 3, 5]}

df_1 = pd.DataFrame(data_1)

plt.pie(df_1['Frequency'], labels = df_1['Music Genre'])
plt.title('The six most popular music genres')
plt.show()
```

```python
film_genre = {}

for datum in my_data:
    film = datum[2]
    if film in film_genre:
        film_genre[film] += 1
    else:
        film_genre[film] = 1

print(film_genre)
# Most popular film genre was Comedy with 7 votes, with Thriller getting 6 votes and Sci-Fi got 5 votes
```

```python
data_2 = {'Film Genre': ['Sci-Fi', 'Adventure', 'Thriller', 'Comedy', 'Cult', 'Action', 'Horror', 'Drama', 'Western', 'Crime', 'Documentary', 'Experimental'], 
          'Frequency': [5, 2, 6, 7, 4, 3, 2, 1, 1, 1, 1, 1]}

df_2 = pd.DataFrame(data_2)

plt.pie(df_2['Frequency'], labels = df_2['Film Genre'])
plt.title('Popularity of Film Genres')
plt.show()
```

```python
art_style = {}

for datum in my_data:
    art = datum[3]
    if art in art_style:
        art_style[art] += 1
    else:
        art_style[art] = 1

print(art_style)
# Most popular art style was Abstract and Surrealism, both with 5 votes, while Pop Art & Digital both have only 2 votes
```

```python
data_3 = {'Art Style': ['Abstract', 'Realism', 'Landscape', 'Surrealism', 'Digital', 'Impressionism', 'Art Noveau', 'Fauvism', 'Pop Art'], 
          'Frequency': [5, 0, 0, 5, 2, 1, 1, 0, 2]}

df_3 = pd.DataFrame(data_3)

plt.pie(df_3['Frequency'], labels = df_3['Art Style'])
plt.title('Popularity of Art Styles')
plt.show()
```

```python
game_genre = {}

for datum in my_data:
    game = datum[4]
    if game in game_genre:
        game_genre[game] += 1
    else:
        game_genre[game] = 1

print(game_genre)
# Most popular game category was Puzzle with 6 votes, while FPS got 5 votes and both RPGs and Horror got 3 votes each.
```
